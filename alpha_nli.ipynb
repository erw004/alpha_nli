```python
# The following is taken from https://github.com/allenai/mosaic-leaderboard/blob/master/anli/baselines/random-baseline/random_baseline.py, which is
# linked in the Getting Started section of the leaderboard https://leaderboard.allenai.org/anli/submissions/get-started

import argparse
import json
from typing import List

# Parse the input file from JSONL to a list of dictionaries.
def read_jsonl_lines(input_file: str) -> List[dict]:
  with open(input_file) as f:
    lines = f.readlines()
    return [json.loads(l.strip()) for l in lines]

# Write the predictions to the output file.
def write_predictions(output_file: str, predicted_answers):
  with open(output_file, "w") as f:
    for p in predicted_answers:
      f.write(p)
      f.write("\n")
      f.close()
```
